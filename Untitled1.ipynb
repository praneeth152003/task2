```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
try:
    titanic_df = pd.read_csv('/content/Titanic-Dataset (2).csv')
except FileNotFoundError:
    print("Error: '/content/Titanic-Dataset (2).csv' not found. Please upload your data file or replace with the correct path.")

    sample_data = {
        'age': [1, 2, 3, 4, 5],
        'fare': [10, 12, 15, 11, 13],
        'class': ['A', 'B', 'A', 'C', 'B']
    }
    titanic_df = pd.DataFrame(sample_data)
    print("Using a sample DataFrame for demonstration.")
print("DataFrame Info:")
titanic_df.info()
print("\nDescriptive Statistics:")
print(titanic_df.describe())
for column in titanic_df.select_dtypes(include='object').columns:
    print(f"\nValue Counts for '{column}':")
    print(titanic_df[column].value_counts())
print("\nCorrelation Matrix:")
print(titanic_df.corr(numeric_only=True))
titanic_df.hist(figsize=(10, 8))
plt.suptitle("Histograms of Numerical Columns")
plt.show()
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic_df.select_dtypes(include=['int64', 'float64']))
plt.title("Box Plots of Numerical Columns")
plt.show()
if 'class' in titanic_df.columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(x='class', data=titanic_df)
    plt.title("Count Plot of 'class'")
    plt.show()
if 'age' in titanic_df.columns and 'fare' in titanic_df.columns:
    fig = px.scatter(titanic_df, x='age', y='fare', title='Scatter Plot of Age vs Fare')
    fig.show()
```
